# WEB SCRAPING NO SITE DO IBGE


```python
#Importando as bibliotecas necessárias

import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#Criando uma função para fazer o resquest e selector

def extraindo_inf(Estado: str) -> dict:
    print(f'Extraindo informações da {Estado} ..')
    estado_url = f'https://www.ibge.gov.br/cidades-e-estados/{Estado}.html'
    page = requests.get(estado_url)
    
    #utilizando os seletores do CSS (indicadores)
    soup = BeautifulSoup(page.content, 'html.parser')
    indicadores = soup.select('.indicador')
    
    
    #Pegando os seletores .ind-label e .ind-value
    inf_do_estado = {
        ind.select('.ind-label')[0].text: ind.select('.ind-value')[0].text
        for ind in indicadores
    }
    inf_do_estado['Estado'] = Estado
    
    
    return inf_do_estado
    

    
extraindo_inf('ba')
```

# Pegando informações de Alguns Estados

```python
#Selecionando os Estados
estados = ['al', 'rn', 'ba', 'sp', 'rj', 'df', 'es', 'go', 'mg', 'pa','pb']

#Armazenando as informações dos Estados
dados_dos_estados = [extraindo_inf(Estado) for Estado in estados]
```

```python
#Transformando as informações dos Estados em uma Tabela
dataframe = pd.DataFrame(dados_dos_estados)

#Selecionando as 06 primeiras linhas (Estados)
dataframe.head(6)
```

```python
# Verificando informações da coluna, percebemos que todas estão (object=string)
dataframe.info()
```

Observação: Para fazermos operações matemáticas e limpezas na tabela é necessário que as colunas sejão int(valor númerico)

# PROCESSO DE LIMPEZA

```python
#Criando uma cópia do meu dataframe
copia_dataframe = dataframe.copy() 
copia_dataframe.head()
```

```python
#Alterando os nomes das colunas
copia_dataframe.columns = ['Governador', 'Capital', 'Gentilico', 'Área', 'População', 'Densidade_demografica', 'Matrículas_Ensino_Fundamental', 'IDH', 'Receitas', 'Despesas', 'Rendimento_per_capita', 'Total_Veiculos']

copia_dataframe.head()
```

```python
#Iremos retirar as ',' e colocar '.' com o metodo replace do pacote pandas

copia_dataframe = copia_dataframe.replace({
        '\.': '',
    ',': '.',
    '\[\d+\]': '',
    ' hab/km²': '',
    ' km²': '',
    ' pessoas': '',
    ' matrículas': '',
    'R\$.*': '',
    ' veículos': ''
}, regex=True)
copia_dataframe.head()
```

# Salvando os dados em CSV 

```python
copia_dataframe.to_csv(index=False)
```
